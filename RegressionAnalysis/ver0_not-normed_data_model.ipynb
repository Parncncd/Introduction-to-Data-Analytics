

## cap the outliers



```python
def cap_outlier(x):
   return x[(x < x.quantile(0.75) + 1.5*(x.quantile(0.75) - x.quantile(0.25))) & (x > x.quantile(0.25) - 1.5*(x.quantile(0.75) - x.quantile(0.25)))]
for i in range(1,9):
   tmp='x'
   tmp +=str(i)
   a_train[tmp] = cap_outlier(a_train[tmp])
   b_train[tmp]=cap_outlier(b_train[tmp])
   a_test[tmp]=cap_outlier(a_test[tmp])
   b_test[tmp]=cap_outlier(b_test[tmp])
a_train['y']=cap_outlier(a_train['y'])
b_train['y']=cap_outlier(b_train['y'])
a_train[['x4']]=cap_outlier(a_train[['x4']])
b_train[['x8']]=cap_outlier(b_train[['x8']])

```

# Train Model

```python
X_a = a_train[["x3","x8"]]
y_a = a_train[["y"]]
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_a, y_a, test_size=0.3, random_state=101)
```

```python
print("len(X_train) : ", len(X_train_a))
print("len(X_test) : ", len(X_test_a))
print("len(y_train) : ", len(y_train_a))
print("len(y_test) : ", len(y_test_a))
```

```python
print(X_train_a)
```

```python
model_a = LinearRegression()
model_a.fit(X_train_a,y_train_a)
```

```python
X_b = b_train[["x1","x2"]]
y_b = b_train["y"]
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_b, y_b, test_size=0.3, random_state=101)
```

```python
model_b = LinearRegression()
model_b.fit(X_train_b,y_train_b)
```

```python
a_y_pred = pd.DataFrame(model_a.predict(X_test_a))
```
