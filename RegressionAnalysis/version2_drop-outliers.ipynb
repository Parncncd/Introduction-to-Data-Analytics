```python
def drop_outlier(x):
   return x[(x < x.quantile(0.75) + 1.5*(x.quantile(0.75) - x.quantile(0.25))) & (x > x.quantile(0.25) - 1.5*(x.quantile(0.75) - x.quantile(0.25)))]
a_train = a_train.apply(drop_outlier)
b_train = b_train.apply(drop_outlier)
```

# Train Model

```python
a_train.fillna(a_train.mean(), inplace=True)
X_a = a_train[["x3","x8"]]
y_a = a_train["y"]
X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_a, y_a, test_size=0.3, random_state=101)
```

```python
print("len(X_train) : ", len(X_train_a))
print("len(X_test) : ", len(X_test_a))
print("len(y_train) : ", len(y_train_a))
print("len(y_test) : ", len(y_test_a))
```

```python
model_a = LinearRegression()
model_a.fit(X_train_a,y_train_a)
```

```python
b_train.fillna(b_train.mean(), inplace=True)
X_b = b_train[["x1","x2"]]
y_b = b_train["y"]
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_b, y_b, test_size=0.25, random_state=101)
```

```python
model_b = LinearRegression()
model_b.fit(X_train_b,y_train_b)
```

## **result** a

```python
y_pred_a = model_a.predict(X_test_a)
y_pred_a
```

## **result b**

```python
y_pred_b= model_b.predict(X_test_b)
y_pred_b
```

# Model evaluation

```python
plt.scatter(y_test_a,y_pred_a)
```

## RMSE_a

```python
rmse = metrics.mean_squared_error(y_test_a, y_pred_a, squared=False)
print("RMSE = ", rmse)
```

```python
model_a.score(X_test_a,y_test_a)
```

## RMSE_b

```python
rmse = metrics.mean_squared_error(y_test_b, y_pred_b, squared=False)
print("RMSE = ", rmse)
```

```python
model_b.score(X_test_b,y_test_b)
```
